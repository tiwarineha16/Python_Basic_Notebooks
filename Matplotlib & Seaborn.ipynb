```python
np.pi
```

```python
x=np.arange(0,4*np.pi,0.1)
y=np.sin(x)
plt.title('Sine Wave Graph')
plt.plot(x,y)
plt.show()
```

```python
x=np.arange(0,4*np.pi,0.1)
y_sin=np.sin(x)
y_cos=np.cos(x)

plt.subplot(2,1,1)
plt.plot(x,y_sin)
plt.title('Sine')
plt.subplot(2,1,2)
plt.plot(x,y_cos)

plt.title('Cosine')
plt.show()
plt.savefig('Test2.png')
```

```python
x1=[2,8,10]
y1=[11,16,19]

x2=[3,9,11]
y2=[6,15,7]

plt.bar(x1,y1)
plt.bar(x2,y2,color='r')
plt.title('Bar Graph')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
```

```python
a=np.array([12,33,44,55,87,66,54,40,324,23,63])
plt.hist(a,bins=20)
plt.title('Histogram')# shows count or density of number on the y axis
plt.show()

```

```python
data=[np.random.normal(0,std,100) for std in range(1,4)]
plt.boxplot(data, vert=True, patch_artist=True)
plt.show()
```

```python
labels='Python','C++','Ruby', 'Java'
sizes=[215, 310, 245, 120]
colors=['gold','yellowgreen','lightcoral','lightskyblue']
explode=(0.1,0,0,0)
plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%', shadow=True)

plt.axis=('equal')
plt.show()
```
